```python
def extract_stipend(stipend):

    if pd.isna(stipend):
        return pd.Series([None, None, None])

    stipend = str(stipend)

    # Extract full Indian-formatted numbers
    numbers = re.findall(r'[\d,]+', stipend)

    cleaned_numbers = []

    for num in numbers:

        try:
            cleaned_numbers.append(
                int(num.replace(",", ""))
            )

        except:
            continue

    if len(cleaned_numbers) == 1:

        value = cleaned_numbers[0]

        return pd.Series([
            value,
            value,
            value
        ])

    elif len(cleaned_numbers) >= 2:

        min_value = cleaned_numbers[0]
        max_value = cleaned_numbers[1]

        avg_value = (
            min_value + max_value
        ) / 2

        return pd.Series([
            min_value,
            max_value,
            avg_value
        ])

    return pd.Series([None, None, None])
```

```python
df[
    [
        "Min_Stipend",
        "Max_Stipend",
        "Avg_Stipend"
    ]
] = df["Stipend"].apply(extract_stipend)
```

```python
def stipend_type(stipend):

    stipend = str(stipend).lower()

    if "unpaid" in stipend:
        return "Unpaid"

    elif "competitive" in stipend:
        return "Competitive"

    elif "nan" in stipend:
        return "Missing"

    return "Paid"
```

```python
def extract_city_workmode(location, remote_type):

    location = str(location)

    remote_type = str(remote_type)

    # Hybrid case
    if "(" in location:

        parts = location.split("(")

        city = parts[0].strip()

        work_mode = (
            parts[1]
            .replace(")", "")
            .strip()
        )

        return pd.Series([city, work_mode])

    # Work from home case
    elif "work from home" in location.lower():

        return pd.Series([
            "Remote",
            "Remote"
        ])

    # Default fallback
    return pd.Series([
        location,
        remote_type
    ])
```

```python
df[
    [
        "City",
        "Work_Mode"
    ]
] = df.apply(
    lambda row:
    extract_city_workmode(
        row["Location"],
        row["Remote_Type"]
    ),
    axis=1
)
```
